# Comparing Adadelta and Adam (+ various LRs)
On CoLA, using BiLSTM (300 hidden, 400 fc units), with 800K GPT-2 augmentation, B=50, word2vec, temperature=3 and MSE loss, over 30 epochs.

```python
line_files = {
                "adadelta": "run-tb_Nov25_15-22-40_landonia11.inf.ed.ac.uk-tag-eval_mcc.csv",
                "adam (lr=5e-2)": "run-tb_Nov26_14-38-26_landonia11.inf.ed.ac.uk-tag-eval_mcc.csv",
                "adam (lr=5e-3)": "run-tb_Nov26_14-44-21_landonia11.inf.ed.ac.uk-tag-eval_mcc.csv",
                "adam (lr=5e-4)": "run-distil-bilstm-CoLA-Nov26-14_37_51_tb_Nov26_14-41-55_landonia11.inf.ed.ac.uk-tag-eval_mcc.csv"
             }
line_data = {label: pd.read_csv(file)["Value"] for label, file in line_files.items()}
```

```python
for label, data in line_data.items():
    plt.plot(data, label=label, lw=0.5)
plt.legend()
plt.show()
```
